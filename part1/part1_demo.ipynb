## 1. Khởi tạo dữ liệu Test Cases


```python
# Import các hàm từ thư mục part1
from gaussian import gaussian_eliminate
from determinant import determinant
from inverse import inverse
from rank_basis import rank_and_basis

print("=== KHỞI TẠO BỘ 5 TEST CASES ===")

# TC 1: Trường hợp cơ bản (Nghiệm duy nhất, ma trận khả nghịch, det != 0)
A1 = [[2, 1, -1], [-3, -1, 2], [-2, 1, 2]]
b1 = [8, -11, -3]

# TC 2: Trường hợp mâu thuẫn (Hệ vô nghiệm)

A2 = [[1, 2, 3], [4, 5, 6], [1, 2, 3]]
b2 = [1, 2, 5] 

# TC 3: Trường hợp phụ thuộc tuyến tính (Hệ vô số nghiệm, ma trận suy biến)

A3 = [[1, 2, 3], [4, 5, 6], [2, 4, 6]]
b3 = [1, 2, 2]

# TC 4: Edge Case - Cần Partial Pivoting (Phần tử chéo A[0][0] = 0)
A4 = [[0, 2, 1], [1, -2, -3], [-1, 1, 2]]
b4 = [5, -1, 3]

# TC 5: Edge Case - Ma trận rỗng (Empty matrix) và Ma trận 1x1
A5_empty = []
b5_empty = []
A5_1x1 = [[5]]
b5_1x1 = [10]

test_cases_A = [A1, A2, A3, A4, A5_empty, A5_1x1]
test_cases_b = [b1, b2, b3, b4, b5_empty, b5_1x1]
names = ["TC1 (Nghiệm duy nhất)", "TC2 (Vô nghiệm)", "TC3 (Vô số nghiệm)", "TC4 (Có số 0 ở đường chéo)", "TC5 (Ma trận rỗng)", "TC6 (Ma trận 1x1)"]
```

## 2. Kiểm thử hàm Giải hệ phương trình (gaussian_eliminate & back_substitution)


```python
print("=== TEST HÀM GAUSSIAN_ELIMINATE & BACK_SUBSTITUTION ===")
for i in range(len(test_cases_A)):
    print(f"\n--- {names[i]} ---")
    try:
        M, x, swaps = gaussian_eliminate(test_cases_A[i], test_cases_b[i])
        print(f"Kết quả nghiệm (x): {x}")
        if type(M) == list and len(M) > 0:
            print(f"Số lần hoán vị dòng: {swaps}")
    except Exception as e:
        print(f"Lỗi/Ngoại lệ bắt được: {e}")
```

## 3. Kiểm thử hàm Tính định thức (determinant)


```python
print("=== TEST HÀM DETERMINANT ===")

test_det_A = [A1, A2, A3, A4, A5_1x1]
det_names = ["TC1 (Cơ bản)", "TC2 (Mâu thuẫn -> Suy biến)", "TC3 (Phụ thuộc -> Suy biến)", "TC4 (Pivoting)", "TC6 (1x1)"]

for i in range(len(test_det_A)):
    print(f"\n--- {det_names[i]} ---")
    try:
        det = determinant(test_det_A[i])
        print(f"Định thức (det A) = {det:.4g}")
    except Exception as e:
        print(f"Lỗi/Ngoại lệ bắt được: {e}")
```

## 4. Kiểm khử ma trận nghịch đảo (INVERSE)

```python
print("=== TEST HÀM INVERSE ===")
for i in range(len(test_det_A)):
    print(f"\n--- {det_names[i]} ---")
    try:
        inv_A = inverse(test_det_A[i])
        if type(inv_A) == str:
            print(inv_A) # In ra thông báo lỗi suy biến của code
        else:
            print("Ma trận nghịch đảo A^-1:")
            for row in inv_A:
                print([round(val, 4) for val in row])
    except Exception as e:
        print(f"Lỗi/Ngoại lệ bắt được: {e}")
```

## 5. Kiểm thử hạng và cơ sở (rank and basis)

```python
print("=== TEST HÀM RANK_AND_BASIS ===")
for i in range(len(test_cases_A)):
    print(f"\n--- {names[i]} ---")
    try:
        res = rank_and_basis(test_cases_A[i])
        if type(res) == str:
            print(res)
        else:
            print(f"Hạng (Rank): {res['rank']}")
            print(f"Cơ sở không gian dòng: {res['basis_row']}")
            print(f"Cơ sở không gian cột: {res['basis_col']}")
            print(f"Cơ sở không gian nghiệm (Null space): {res['basis_null']}")
    except Exception as e:
        print(f"Lỗi/Ngoại lệ bắt được: {e}")
```

```python
import numpy as np

def verify_solution_logic(A, x, b):
    if not isinstance(x, list): 
        # Nếu x là chuỗi báo lỗi, kiểm tra xem NumPy có xác nhận hệ đó đặc biệt không
        return "True (Hệ đặc biệt)"
    return np.allclose(np.array(A) @ np.array(x), np.array(b), atol=1e-5)

def verify_inverse(A, A_inv):
    if not isinstance(A_inv, list): 
        return "True (Ma trận suy biến)"
    A_np = np.array(A)
    inv_np = np.array(A_inv)
    I = np.eye(len(A_np))
    return np.allclose(A_np @ inv_np, I, atol=1e-5)

def verify_determinant(A, det_custom):
    det_np = np.linalg.det(np.array(A))
    return np.allclose(det_custom, det_np, atol=1e-5)

def verify_rank(A, rank_custom):
    rank_np = np.linalg.matrix_rank(np.array(A))
    return rank_custom == rank_np

# --- 2. HÀM CHÍNH: verify_solution(A, x, b) 

def verify_solution(A, x, b):
    # In ma trận A 
    print("Ma trận A:")
    for row in A:
        print(f"  {row}")

    # Lấy các giá trị từ hàm code tay 
    det_val = determinant(A)
    inv_val = inverse(A)
    rb_val = rank_and_basis(A)

    # In kết quả kiểm chứng kiểu True/False
    print("\nKết quả kiểm chứng:")
    print(f"solution: {verify_solution_logic(A, x, b)}")
    print(f"inverse:  {verify_inverse(A, inv_val)}")
    print(f"det:      {verify_determinant(A, det_val)}")
    print(f"rank:     {verify_rank(A, rb_val['rank'])}")

# --- 3. VÒNG LẶP CHẠY FULL TEST CASES
print("===== KIỂM CHỨNG CHI TIẾT =====")


for i in range(len(test_cases_A)):
    print(f"\n-{names[i]}")
    
    # Xử lý trường hợp ma trận rỗng để không bị văng lỗi
    if not test_cases_A[i] or (len(test_cases_A[i]) > 0 and not test_cases_A[i][0]):
        print("Trạng thái: Bỏ qua do ma trận rỗng.")
        print("-" * 40)
        continue
        
    try:
        # Lấy nghiệm x từ hàm gaussian_eliminate 
        _, x_ans, _ = gaussian_eliminate(test_cases_A[i], test_cases_b[i])
        
        # Gọi hàm verify chính
        verify_solution(test_cases_A[i], x_ans, test_cases_b[i])
    except Exception as e:
        print(f"Lỗi khi xử lý bộ test này: {e}")
        
    print("-" * 40)
```
